```python
# Import dependencies
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats import linregress
from matplotlib import pyplot as plt
from pathlib import Path
```

```python
# Read in CSV
datafile = Path("predicting_misuse_csv.csv")
original_df = pd.read_csv(datafile)
original_df.head()
```

```python
# Add a column for sum
original_df["Drug Use Total"] = ""
original_df.head()
```

```python
# Sum illict drug use (without Rx misuse) and add to empty column (function from chatGPT)
selected_columns = ["HEROINUSE", "TRQLZRS","SEDATVS","COCAINE","AMPHETMN","HALUCNG"]
for index, row in original_df.iterrows():
    original_df.loc[index, 'Drug Use Total'] = row[selected_columns].sum()

original_df.head()
```

```python
# Create a correlation matrix to determine which variables are more highly correlated with illicit drug use  (https://www.geeksforgeeks.org/create-a-correlation-matrix-using-python/)
matrix = original_df.corr()
plt.imshow(matrix, cmap='Blues')
plt.colorbar()
variables = []
for i in matrix.columns:
  variables.append(i)
plt.xticks(range(len(matrix)), variables, rotation="vertical", ha='center')
plt.yticks(range(len(matrix)), variables)
plt.show()
print(matrix)
```

```python
# Create new dataframe to compare drug use to education
# Rename columns
education_df = original_df [["YEAR", "EDUCAT", "HEROINUSE", "TRQLZRS", "SEDATVS", "COCAINE", "AMPHETMN", "HALUCNG", "Drug Use Total"]]
education_df = education_df.rename(columns={"YEAR": "Year", 
                                            "EDUCAT": "Education", 
                                            "HEROINUSE": "Heroin Use", 
                                            "TRQLZRS": "Tranquilizer Use", 
                                            "SEDATVS": "Sedative Use", 
                                            "COCAINE": "Cocaine Use", 
                                            "AMPHETMN": "Amphetamine Use", 
                                            "HALUCNG": "Hallucinogen Use"
                                           })
education_df.head()
```

```python
education_df['Drug Use Total'] = education_df['Drug Use Total'].astype(np.int64)
education_df.dtypes
```

```python
# Export the Education dataframe into a csv
education_df.to_csv("output_data/education.csv", index_label="City_ID")
```

```python
x_values = education_df["Education"]
y_values = education_df["Drug Use Total"]
# (slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
print(x_values.isnull().sum())
print(y_values.isnull().sum())
```

```python
# Define a function to create Linear Regression plots
def Lin_Reg_Plt(df, x, y):
    x_values = df[x]
    y_values = df[y]
    
    (slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
    rvalue = round(rvalue, 5)
    
    regress_values = x_values * slope + intercept
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
    
    plt.scatter(x_values, y_values, s=60, color='darkcyan', edgecolor='black')
    
    plt.plot(x_values,regress_values,"r-")
    plt.annotate(line_eq, xy=(0.5, 0.1), xycoords='axes fraction', fontsize=15,color="red", ha='center', va='bottom')

    plt.xlabel(x)
    plt.ylabel(y)
    print("The r-value is: " + str(rvalue))
    plt.show()
```

```python
Lin_Reg_Plt(education_df ,"Education", "Drug Use Total")
```
