```python
# Import dependencies
from pathlib import Path
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
# Read in CSV
datafile = Path("worldometer_coronavirus_daily_data.csv")
daily_covid_df = pd.read_csv(datafile)

# Convert date coulumn to datetime and create a 'year_month' column
daily_covid_df['date'] = pd.to_datetime(daily_covid_df['date'])
daily_covid_df['year_month'] = daily_covid_df['date'].dt.strftime('%Y-%m')
daily_covid_df.head()
```

```python
# Convert the daily data to Monthly and create new dataframe
# Sum the 'daily_new_cases' of each month for each country to get 'monthly_new_cases'
monthly_new_cases = daily_covid_df.groupby(['country', 'year_month'])['daily_new_cases'].sum().reset_index()
monthly_new_cases = monthly_new_cases.rename(columns={'daily_new_cases': 'monthly_new_cases'})

# Sum the 'daily_new_deaths' of each month for each country to 'get monthly_new_deaths'
monthly_new_deaths = daily_covid_df.groupby(['country', 'year_month'])['daily_new_deaths'].sum().reset_index()
monthly_new_deaths = monthly_new_deaths.rename(columns={'daily_new_deaths': 'monthly_new_deaths'})

# Group by 'year_month' then filter to select the last day of each month
last_day_of_month = daily_covid_df.groupby(['year_month']).apply(lambda x: x[x['date'] == x['date'].max()])

# Reset index to remove multi-index resulting from groupby
last_day_of_month.reset_index(drop=True, inplace=True)

# Remove the daily columns from 'last_day_of_month'
last_day_of_month = last_day_of_month.drop(columns=['daily_new_cases', 'daily_new_deaths'])

# Merge the monthly sums data into 'last_day_of_month' to create 'monthly_covid_df'
monthly_covid_df = pd.merge(last_day_of_month, monthly_new_cases, on=['country', 'year_month'], how='left')
monthly_covid_df = pd.merge(monthly_covid_df, monthly_new_deaths, on=['country', 'year_month'], how='left')

# Reorganize the column order
monthly_covid_df = monthly_covid_df[['year_month', 'country', 'cumulative_total_cases', 'monthly_new_cases', 
                                    'active_cases', 'cumulative_total_deaths', 'monthly_new_deaths']]

# Print the resulting DataFrame
monthly_covid_df.head()
```

```python
# Groupby 'year_month' and 'daily_new_cases' to plot the average new cases over time
total_new_cases_df = monthly_covid_df.groupby('year_month')['monthly_new_cases'].sum().reset_index()
plt.figure(figsize=(10, 6))
plt.plot(total_new_cases_df['year_month'], total_new_cases_df['monthly_new_cases'], marker='o', linestyle='-')
plt.title('Global COVID Cases')
plt.ylabel('Total Monthly New Cases')
plt.xticks(rotation=45)  
plt.show()
```

```python
# Groupby 'year_month' and 'active_cases' to plot the average active cases over time
active_cases_df = monthly_covid_df.groupby('year_month')['active_cases'].sum().reset_index()
plt.figure(figsize=(10, 6))
plt.plot(active_cases_df['year_month'], active_cases_df['active_cases'], marker='o', linestyle='-')
plt.title('Global COVID Cases')
plt.ylabel('Total Active Cases')
plt.xticks(rotation=45)  
plt.show()
```

```python
total_deaths_df = monthly_covid_df.groupby('year_month')['monthly_new_deaths'].sum().reset_index()
plt.figure(figsize=(10, 6))
plt.plot(total_deaths_df['year_month'], total_deaths_df['monthly_new_deaths'], marker='o', linestyle='-')
plt.title('Global COVID Deaths')
plt.ylabel('Total Monthly Deaths')
plt.xticks(rotation=45)  
plt.show()
```

```python
# Merge the average new cases, active case and deaths dataframes
month_pivot_df = pd.merge(pd.merge(total_new_cases_df, active_cases_df, on=['year_month']), total_deaths_df, on='year_month')
month_pivot_df.set_index('year_month', inplace=True)
# month_pivot_df.head()
```

```python
# Plot each column of month_pivot_df on individual graphs
plt.figure(figsize=(10, 10))
for i, column in enumerate(month_pivot_df.columns):
    ax = plt.subplot(len(month_pivot_df.columns), 1, i + 1)
    ax.plot(month_pivot_df.index, month_pivot_df[column], marker='o', linestyle='-', label=column)
    ax.set_title(column)
    ax.set_ylabel('Total Monthly')
    plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 8))

for column in month_pivot_df.columns:
    plt.plot(month_pivot_df.index, month_pivot_df[column], marker='o', linestyle='-', label=column)


plt.title('Total Global COVID')
plt.ylabel('Monthly Total')
plt.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Normalize the data in each column
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(month_pivot_df)
monthly_normalized_df = pd.DataFrame(scaled_data, columns=month_pivot_df.columns, index=month_pivot_df.index)

# Plot the normalized data
plt.figure(figsize=(15, 8))
for column in monthly_normalized_df.columns:
    plt.plot(monthly_normalized_df.index, monthly_normalized_df[column], marker='o', linestyle='-', label=column)

plt.title('Global COVID (Normalized)')
plt.ylabel('Total Monthly (Normalized)')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
